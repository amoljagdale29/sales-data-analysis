```python
import pandas as pd
import glob
import itertools
```

```python
# import csv's from folder and merge 
path = 'Sales_Data'
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True)
frame
```

### Clean data

```python
frame=frame.dropna()
```

```python
# format and make date index
frame['temp_date'] = pd.to_datetime(frame['Order Date'],format="%m/%d/%y %H:%M", errors='coerce').isna()
frame[frame['temp_date']==True]
#data contains multiple column header 355 rows
```

```python
frame = frame[frame['temp_date']==False]
frame = frame.drop(['temp_date'],axis=1)
```

```python
frame['Order Date(formated)'] = pd.to_datetime(frame['Order Date'],format="%m/%d/%y %H:%M")
frame
```

```python
# add order price column for total sales
frame['Quantity Ordered'] = frame['Quantity Ordered'].astype('float32')
frame['Price Each'] = frame['Price Each'].astype('float32')
frame['Order Price'] = frame['Quantity Ordered'] * frame['Price Each']
frame
```

### which is best sales month, how much sales happend in that month?

```python
# group by month on total data and find sum of 'order price'
grouped_data = frame.copy()
grouped_data = grouped_data.set_index('Order Date(formated)')
GB=grouped_data.groupby(pd.Grouper(freq="M")).sum()
```

```python
GB.plot.bar(y='Order Price')
```

## Ans: best sales in dec 2019 with 4.16x10^6 

## Which city had best sales?

```python
# split address in street,city,state and then group by city+state
grouped_data = frame.copy()
grouped_data = grouped_data.set_index('Order Date(formated)')
grouped_data[['street','city','state']] = grouped_data['Purchase Address'].str.split(',', expand=True)
GB=grouped_data.groupby(['city','state']).sum()
GB
```

```python
GB.plot.bar(y='Order Price')
```

### Ans : San Francisco had best sales with 8.26x10^6

## What time should we display advertisemens to maximize the likelihood of customerâ€™s buying product?

```python
# get time from formated date and group by hour 
grouped_data = frame.copy()
grouped_data = grouped_data.set_index('Order Date(formated)')
GB=grouped_data.groupby([grouped_data.index.hour]).count()
GB
```

```python
GB.plot.bar(y='Order Price')
```

### Ans : Morning 10 to 1 pm and evening 5pm to 8pm would be ideal time to advertise

## Which are most bought item together?

```python
grouped_data = frame.copy()
grouped_data = grouped_data.set_index('Order Date(formated)')
v = grouped_data['Order ID'].value_counts()
# get rows where order id is repeated
grouped_data = grouped_data[grouped_data['Order ID'].isin(v.index[v.gt(1)])]
grouped_data
```

```python
# function for getting list of products in one order
def get_same_product(x):
    temp  = grouped_data[grouped_data['Order ID']==x]
    output = []
    for index, row in temp.iterrows():
        if row['Product'] not in output:
            output.append(row['Product'])
    return x,output
order_ids = grouped_data.copy()['Order ID']
order_ids = order_ids.drop_duplicates(keep='last')
order_ids = pd.DataFrame({'Order Date(formated)':order_ids.index, 'Order ID':order_ids.values})
grouped_products= order_ids.apply(lambda x : get_same_product(x['Order ID']),axis=1,result_type='expand')
# # grouped_products[['order_id','product_list']] = grouped_data[:10:].apply(lambda x : get_same_product(x['Order ID']),axis=1,result_type='expand')
grouped_products
```

```python
# Getting unique products and making combination of that
unique_product = grouped_data['Product'].unique()
unique_product.tolist()
unique_combo_list = []
import itertools
for L in range(2, 4):
    for subset in itertools.combinations(unique_product, L):
        unique_combo_list.append(list(subset))
print(len(unique_combo_list))
print(unique_combo_list[:10])
```

```python
grouped_products = pd.DataFrame({'Order Date(formated)':grouped_products.index, 'Products':grouped_products[1].values})
```

```python
# getting counts of each combination of unique products in orders
grouped_products
def count_product(df):
    output = [0 for i in range(len(unique_combo_list))]
    for i,row in df.iterrows():
        for k,prod_combo in enumerate(unique_combo_list):
            if set(prod_combo) <= set(row['Products']):
                output[k] = output[k]+1
    return output
counts = count_product(grouped_products)
counts
```

```python
output_prod_combo = pd.DataFrame(list(zip(unique_combo_list,counts)),columns=['combo','count'])
```

```python
# getting sorting values above threshold
threshold = 100
ploting_df = output_prod_combo[output_prod_combo['count']>=threshold].sort_values('count') #threshold count
ploting_df
```
